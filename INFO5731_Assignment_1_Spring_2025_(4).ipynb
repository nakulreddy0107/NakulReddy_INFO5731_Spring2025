$ V=\frac{2}{3}\ast\ \pi\ast\ r^3 $

	r	V

# Question 3 (10 points).

Create a Python program that encompasses various string manipulation task:

(a) Prompt the user to input two sentences.

(b) Transform each sentence into a list, where each element corresponds to a word in the sentence. For example, if the sentences are "Hello, how are you doing today?" and "This is an example sentence for the program!"

The program should output:
['Hello', ',', 'how', 'are', 'you', 'doing', 'today', '?', 'This', 'is', 'an', 'example', 'sentence', 'for', 'the', 'program', '!']

(c) Compare the two sentences. Print out a list of words that occur in both sentences.

(d) Remove any punctuation from the sentence and print the sanitized sentences. It should remove at least the following characters, but it can remove more: period(“.”), comma (“,”), semicolon (“;”), and colon (“:”).

(e) Count the number of vowels in each of the sentences. Print out each vowel and the occurrences  in the each of the two sentences separately, such as:

Sentence1: a:2, e:2, i:0, etc.
Sentence2: a:1, e:2, i:2, etc.

Design a well-structured Python program to execute these tasks efficiently.

```python
# Your code here
import string
from collections import Counter

def get_sentences():
    sentence1 = input("Enter the first sentence: ")
    sentence2 = input("Enter the second sentence: ")
    return sentence1, sentence2

def sentence_to_list(sentence):
    return sentence.split()

def find_common_words(list1, list2):
    return list(set(list1) & set(list2))

def remove_punctuation(sentence):
    return ''.join(ch for ch in sentence if ch not in string.punctuation)

def count_vowels(sentence):
    vowels = "aeiouAEIOU"
    sentence_vowels = Counter(ch.lower() for ch in sentence if ch.lower() in vowels)
    return {v: sentence_vowels.get(v, 0) for v in "aeiou"}

def main():
    sentence1, sentence2 = get_sentences()

    word_list1 = sentence_to_list(sentence1)
    word_list2 = sentence_to_list(sentence2)

    combined_list = word_list1 + word_list2
    print("Word List:", combined_list)

    common_words = find_common_words(word_list1, word_list2)
    print("Common Words:", common_words)

    clean_sentence1 = remove_punctuation(sentence1)
    clean_sentence2 = remove_punctuation(sentence2)
    print("Sanitized Sentence 1:", clean_sentence1)
    print("Sanitized Sentence 2:", clean_sentence2)

    vowel_count1 = count_vowels(clean_sentence1)
    vowel_count2 = count_vowels(clean_sentence2)

    print("Vowel Count in Sentence 1:", ", ".join(f"{k}:{v}" for k, v in vowel_count1.items()))
    print("Vowel Count in Sentence 2:", ", ".join(f"{k}:{v}" for k, v in vowel_count2.items()))

if __name__ == "__main__":
    main()

```

```python
import math

def  mathematical_expression(expression):
    try:
        result = eval(expression)
        return result
    except ZeroDivisionError:
        return "Error: Division by zero is not allowed."
    except Exception as e:
        return f"Error: {str(e)}"

def calculator():
    print("Supported operations: +, -, *, /, (), ** (exponentiation), sqrt (square root), % (modulus)")
    print("Type 'exit' to quit.")

    while True:
        expression = input("Enter a mathematical expression: ")
        if expression.lower() == 'exit':
            break
        expression = expression.replace('sqrt', 'math.sqrt')
        result = mathematical_expression(expression)
        print(f"Result: {result}")
calculator()
```

# Question 6 (20 points).
Develop a advance Python program to manage inventory records for a prodcut managing company. The system needs functions for adding, deleting and changing products and basic search options. Your system should use CSV files to automate restocking and produce reports that show item values alongside category breakdowns. Manage edge cases and corrupted CSV files. Finally maintaine 50 products of different categories in inventory.

```python
# Your code hereimport csv
import os
import csv
import random
import string

# Constants
FILENAME = 'inventory.csv'
FIELDS = ['ID', 'Name', 'Category', 'Price', 'Quantity']


def initialize_inventory():
    # Check if the inventory file exists, if not, create it with headers
    if not os.path.exists(FILENAME):
        with open(FILENAME, mode='w', newline='') as file:
            writer = csv.DictWriter(file, fieldnames=FIELDS)
            writer.writeheader()


def load_inventory():
    inventory = []
    with open(FILENAME, mode='r') as file:
        reader = csv.DictReader(file)
        for row in reader:
            try:
                row['Price'] = float(row.get('Price', 0))
                row['Quantity'] = int(row.get('Quantity', 0))
                # Ensure ID is unique and convert data types
                row['ID'] = int(row.get('ID', 0))
                inventory.append(row)
            except ValueError:
                print(f"Error loading product with ID {row.get('ID')}: Incorrect data format.")
    return inventory


def save_inventory(inventory):
    with open(FILENAME, mode='w', newline='') as file:
        writer = csv.DictWriter(file, fieldnames=FIELDS)
        writer.writeheader()
        writer.writerows(inventory)


def add_product(product):
    inventory = load_inventory()
    if any(item['ID'] == product['ID'] for item in inventory):
        print("Product ID already exists. Please use a unique ID.")
        return
    inventory.append(product)
    save_inventory(inventory)
    print(f"Product {product['Name']} added successfully.")


def delete_product(product_id):
    inventory = load_inventory()
    filtered_inventory = [product for product in inventory if product['ID'] != product_id]
    if len(inventory) == len(filtered_inventory):
        print("Product ID not found.")
    else:
        save_inventory(filtered_inventory)
        print(f"Product with ID {product_id} deleted successfully.")


def update_product(product_id, updated_info):
    inventory = load_inventory()
    for product in inventory:
        if product['ID'] == product_id:
            product.update(updated_info)
            save_inventory(inventory)
            print(f"Product with ID {product_id} updated successfully.")
            return
    print("Product ID not found.")


def search_product(name=None, category=None):
    inventory = load_inventory()
    results = []
    for product in inventory:
        if name and name.lower() in product['Name'].lower():
            results.append(product)
        elif category and category.lower() in product['Category'].lower():
            results.append(product)
    return results


def generate_reports():
    inventory = load_inventory()
    category_breakdown = {}
    total_value = 0
    for product in inventory:
        category = product['Category']
        value = product['Price'] * product['Quantity']
        total_value += value
        if category in category_breakdown:
            category_breakdown[category] += value
        else:
            category_breakdown[category] = value

    print("Category Breakdown:")
    for category, value in category_breakdown.items():
        print(f"Category: {category}, Total Value: {value}")
    print(f"Total Inventory Value: {total_value}")

# Initialize the inventory file
initialize_inventory()

# Adding initial 50 products and trying some edge cases

def generate_random_product(id):
    # Randomly generate a product name
    product_name = 'Product' + ''.join(random.choices(string.ascii_uppercase, k=5))

    # Randomly select a category (e.g., Category1, Category2, etc.)
    category = 'Category' + str(random.randint(1, 5))

    # Randomly generate a price between 5 and 500
    price = round(random.uniform(5.0, 500.0), 2)

    # Randomly generate a quantity between 1 and 100
    quantity = random.randint(1, 100)

    return {
        "ID": id,
        "Name": product_name,
        "Category": category,
        "Price": price,
        "Quantity": quantity
    }

def generate_and_add_products():
    # Generate and add 50 random products
    for i in range(1, 51):
        random_product = generate_random_product(i)
        add_product(random_product)

# Initialize the inventory file
initialize_inventory()

# Generate and add random products
generate_and_add_products()

# You can now also generate reports or perform other operations like deletion, updates, etc.
generate_reports()
# Testing edge cases
add_product({"ID": 1, "Name": "DuplicateProduct", "Category": "Category2", "Price": 15.0, "Quantity": 10})
delete_product(999)  # Non-existent product
update_product(999, {"Name": "UpdatedName"})  # Update non-existent product
search_product(name="NonExistentProduct")
```

# Question 7 (20 points)

Create a well desined Python program for a flight booking system. Offer core functions for adding new flights, finding trips, reserving and releasing seating space and creating performance reports. Ensure the system manages edge cases like invalid input, overbooking, and cancellations. Finally, extend the booking system by including advanced features and maintain about 50 flight in the system.

```python
import random

class Flight:
    def __init__(self, flight_id, origin, destination, seats):
        self.flight_id = flight_id
        self.origin = origin
        self.destination = destination
        self.total_seats = seats
        self.available_seats = seats
        self.reservations = {}

    def reserve_seat(self, passenger_name):
        if self.available_seats > 0:
            if passenger_name in self.reservations:
                return f"Passenger {passenger_name} already has a reservation."
            seat_number = self.total_seats - self.available_seats + 1
            self.reservations[passenger_name] = seat_number
            self.available_seats -= 1
            return f"Seat {seat_number} reserved for {passenger_name} on flight {self.flight_id}."
        else:
            return "No available seats on this flight. Overbooking is not allowed."

    def release_seat(self, passenger_name):
        if passenger_name in self.reservations:
            self.available_seats += 1
            del self.reservations[passenger_name]
            return f"Reservation for {passenger_name} has been cancelled."
        else:
            return "No reservation found for the given passenger."

    def __str__(self):
        return f"Flight {self.flight_id} ({self.origin} -> {self.destination}) - Available Seats: {self.available_seats}/{self.total_seats}"


class FlightBookingSystem:
    def __init__(self):
        self.flights = {}
        self.initialize_flights()

    def initialize_flights(self):
        origins = ["New York", "Los Angeles", "Chicago", "Dallas", "Miami"]
        destinations = ["London", "Tokyo", "Paris", "Berlin", "Sydney"]
        for i in range(1, 51):
            origin = random.choice(origins)
            destination = random.choice(destinations)
            while origin == destination:
                destination = random.choice(destinations)
            seats = random.randint(50, 200)
            flight = Flight(f"F{i:03}", origin, destination, seats)
            self.flights[flight.flight_id] = flight

    def find_flight(self, origin, destination):
        return [flight for flight in self.flights.values() if flight.origin == origin and flight.destination == destination]

    def book_seat(self, flight_id, passenger_name):

        if flight_id in self.flights.keys():
            return self.flights[flight_id].reserve_seat(passenger_name)
        else:

            return "Invalid flight ID. Please enter a valid flight number."

    def cancel_reservation(self, flight_id, passenger_name):
        if flight_id in self.flights:
            return self.flights[flight_id].release_seat(passenger_name)
        else:
            return "Invalid flight ID. Please enter a valid flight number."

    def generate_report(self):
        report = "Flight Booking System Report:\n"
        for flight in self.flights.values():
            report += str(flight) + "\n"
        return report

# Example Usage
system = FlightBookingSystem()
print(system.generate_report())

# Valid Booking and Cancellation

print(system.cancel_reservation("F001", "Alice"))

# Invalid Flight ID
print(system.book_seat("dvb", "Bob"))  # Should return invalid flight ID error
print(system.cancel_reservation("sam", "Charlie"))  # Should return invalid flight ID error

# Overbooking Scenario
for _ in range(210):  # Attempt to overbook beyond capacity
    print(system.book_seat("F001", f"Passenger{_}"))

# Duplicate Booking Attempt
  # Should return duplicate reservation error

# Cancellation for Non-existent Passenger
print(system.cancel_reservation("F002", "Unknown"))  # Should return no reservation found
```
